# Effects of the COVID-19 Pandemic on Student Performance
COVID affected much of university for thousands of students. Zoom lectures, online tests, digitalized homework -- the collegiate experience became very different after 2020. With this shift, many people have come to believe that college was easier during the pandemic, claiming students achieved higher grade point averages. Is this true? Did students truly fair better from a grade point perspective during the pandemic than they did before?

**Central Question:** Did online learning positively skew student GPA's during the COVID-19 pandemic?  
**Hypothesis:** Due to the lack of accountability most students faced, we believe that students did perform better during the pandemic than they did before.  
**Dataset:** https://udc.vt.edu/irdata/data/courses/grades  

**Methodology:** Using the GPA data from Virginia Tech, we can examine (on the aggregate as well as on a course-by-course basis) the difference between average GPA during pre-pandemic (before Spring 2020), pandemic (before Spring 2021, after Spring 2020), and post-pandemic (after Spring 2021) semesters. We can then graph these splits to examine the differences in GPA between the time periods. The average GPA will be taken using the course/time period mean. Additionally, we can evaluate how different student demographics' academic experiences were impacted throughout the pandemic using a similar measure.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 2015-2023 Course Data
courses = pd.read_csv('CourseData.csv')
courses
```

```python
# Convert the 'Academic Year' column to datetime format
courses['Academic Year'] = pd.to_datetime(courses['Academic Year'], format='%Y-%y')

# Create a datetime object for Spring 2020 and Spring 2021
spring_2020 = pd.to_datetime('2020-01-01')
spring_2021 = pd.to_datetime('2021-01-01')

# Split the dataframe into three dataframes based on academic year and term
courses_before_covid = courses.loc[((courses['Academic Year'] < spring_2020) | 
                            ((courses['Academic Year'] == spring_2020) & (courses['Term'] == 'Fall'))) & 
                            (courses['Term'].isin(['Spring', 'Fall']))]

courses_during_covid = courses.loc[((courses['Academic Year'] >= spring_2020) & (courses['Academic Year'] <= spring_2021)) & 
                            (courses['Term'].isin(['Spring', 'Fall']))]

courses_after_covid = courses.loc[((courses['Academic Year'] > spring_2021) | 
                            ((courses['Academic Year'] == spring_2021) & (courses['Term'] == 'Fall'))) & 
                            (courses['Term'].isin(['Spring', 'Fall']))]

# Print the first few rows of each dataframe to verify the split
print(courses_before_covid.head())
print(courses_during_covid.head())
print(courses_after_covid.head())
```

```python
# Calculate the average GPA for each period
avg_gpa_before = courses_before_covid['GPA'].mean()
avg_gpa_during = courses_during_covid['GPA'].mean()
avg_gpa_after = courses_after_covid['GPA'].mean()

# Plot the average GPA for each period
x_labels = ['Before COVID', 'During COVID', 'After COVID']
y_values = [avg_gpa_before, avg_gpa_during, avg_gpa_after]

# Set the style for the plot
sns.set(style="whitegrid")

# Create a line plot with custom markers and colors
fig, ax = plt.subplots()
line_plot = sns.lineplot(x=x_labels, y=y_values, marker='o', markersize=10, linewidth=2, color='blue')

# Customize the plot appearance
ax.set_ylabel('Average GPA', fontsize=14)
ax.set_title('Average GPA Before, During, and After COVID', fontsize=16)
ax.set_ylim(np.min(y_values) - 0.1, np.max(y_values) + 0.1)  # Set y-axis limits

# Annotate data points with their values
for i, value in enumerate(y_values):
    ax.text(i, value + 0.01, f'{value:.2f}', ha='center', fontsize=12, color='black')

# Remove top and right spines for a cleaner look
sns.despine()

plt.show()
```

# An Answer on the Aggregate

We can see that there is a slight difference in academic achievement between the before/after COVID columns and the during COVID column (a difference of 0.1 and 0.7, respectively). There are multiple different ways in which GPA's can shift -- was this caused by a higher frequency of A's, or did the typical ratios of GPA's all shift up during COVID?

```python

# Calculate the average percentage for each grade category
grade_categories = ['A (%)', 'A- (%)', 'B+ (%)', 'B (%)', 'B- (%)', 'C+ (%)', 'C (%)', 'C- (%)', 'D+ (%)', 'D (%)', 'D- (%)', 'F (%)']
grade_distribution_before = courses_before_covid[grade_categories].mean()
grade_distribution_during = courses_during_covid[grade_categories].mean()
grade_distribution_after = courses_after_covid[grade_categories].mean()

# Prepare data for the grouped bar chart
data = pd.DataFrame({'Grade Category': np.tile(grade_categories, 3),
                     'Percentage': np.concatenate([grade_distribution_before, grade_distribution_during, grade_distribution_after]),
                     'Period': np.repeat(['Before COVID', 'During COVID', 'After COVID'], len(grade_categories))})

# Set the style for the plot
sns.set(style="whitegrid")

# Create a grouped bar chart with custom colors
palette = sns.color_palette("muted", 3)
fig, ax = plt.subplots(figsize=(12, 6))
bar_plot = sns.barplot(x='Grade Category', y='Percentage', hue='Period', data=data, palette=palette, errorbar=None)

# Customize the plot appearance
ax.set_ylabel('Percentage', fontsize=14)
ax.set_title('Grade Distribution Before, During, and After COVID', fontsize=16)
ax.set_xticklabels(grade_categories, rotation=45, ha='right', fontsize=12)
ax.legend(title='Period', title_fontsize=14, fontsize=12)

# Remove top and right spines for a cleaner look
sns.despine()

plt.show()
```

# Increase in A's

It appears that we have our answer to our previous question: students on the whole achieved A's at a much higher rate during COVID than they did before and after. Something interesting to note, however, is that after COVID, the ratio of student's achieving A's did not return to pre-pandemic levels.

We can see that students did perform better on a GPA basis during the pandemic than they did before and after. That only tells part of the story though. Course withdrawals and drops can both skew the results of a class' GPA. Was the increase in GPA during COVID due to an increase in withdrawals?

```python
courses_before_covid = courses_before_covid.copy()
courses_during_covid = courses_during_covid.copy()
courses_after_covid = courses_after_covid.copy()

# Calculate the withdraw rate for each period
courses_before_covid.loc[:, 'Withdraw Rate'] = courses_before_covid['Withdraws'] / courses_before_covid['Graded Enrollment']
courses_during_covid.loc[:, 'Withdraw Rate'] = courses_during_covid['Withdraws'] / courses_during_covid['Graded Enrollment']
courses_after_covid.loc[:, 'Withdraw Rate'] = courses_after_covid['Withdraws'] / courses_after_covid['Graded Enrollment']

# Calculate the average withdraw rate for each period
avg_withdraw_rate_before = courses_before_covid['Withdraw Rate'].mean()
avg_withdraw_rate_during = courses_during_covid['Withdraw Rate'].mean()
avg_withdraw_rate_after = courses_after_covid['Withdraw Rate'].mean()

# Plot the average withdraw rate for each period using an area plot
x_labels = ['Before COVID', 'During COVID', 'After COVID']
y_values = [avg_withdraw_rate_before, avg_withdraw_rate_during, avg_withdraw_rate_after]

# Create a DataFrame for the area plot
data = pd.DataFrame({'Period': x_labels, 'Average Withdraw Rate': y_values})

# Create the area plot
fig, ax = plt.subplots()
data.plot.area(x='Period', y='Average Withdraw Rate', ax=ax, alpha=0.3, legend=None)

# Customize the plot appearance
ax.set_ylabel('Average Withdraw Rate', fontsize=14)
ax.set_title('Average Withdraw Rate Before, During, and After COVID', fontsize=16)

# Remove top and right spines for a cleaner look
sns.despine()

plt.show()
```

# Fewer Withdrawals

As we can see, the spike in GPA during COVID was likely not due to an increase in course withdrawals. In fact, it appears that students were more likely to remain in a course during COVID than they were before/after.

Seeing that withdrawals were not a major factor, we need to examine other possibilities for why GPA increased during the COVID-19 pandemic. Could it have been that students that were barely interested in their coursework chose to not return to Virginia Tech during online learning, thus changing the overall student population to those more motivated? To test this, we should first examine the average enrollment per period.

```python
# Calculate the average enrollment for each period
avg_enrollment_before = courses_before_covid['Graded Enrollment'].mean()
avg_enrollment_during = courses_during_covid['Graded Enrollment'].mean()
avg_enrollment_after = courses_after_covid['Graded Enrollment'].mean()

# Plot the average enrollment for each period
x_labels = ['Before COVID', 'During COVID', 'After COVID']
y_values = [avg_enrollment_before, avg_enrollment_during, avg_enrollment_after]

# Set the style for the plot
sns.set(style="whitegrid")

# Create a scatter plot with custom markers and colors
fig, ax = plt.subplots()
scatter_plot = sns.scatterplot(x=x_labels, y=y_values, marker='o', s=100, color='blue')

# Customize the plot appearance
ax.set_ylabel('Average Enrollment', fontsize=14)
ax.set_title('Average Enrollment Before, During, and After COVID', fontsize=16)
ax.set_ylim(np.min(y_values) - 5, np.max(y_values) + 5)  # Set y-axis limits

# Annotate data points with their values
for i, value in enumerate(y_values):
    ax.text(i, value + 1, f'{value:.1f}', ha='center', fontsize=12, color='black')

# Remove top and right spines for a cleaner look
sns.despine()

plt.show()
```

# A Dip in Enrollment

It does appear that there was a dip in enrollment during the COVID-19 pandemic, but it was very slight, and was rapidly remedied after the pandemic. Seeing that this was likely not a factor, we need to think about what else could have happened to increase student GPA outside of our hypothesis of the nature of online work.

Could it have been that students took fewer high course-credit classes, and thus had less demanded from them per class?

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a stacked histogram
fig, ax = plt.subplots()
bins = np.arange(0, courses_before_covid['Credits'].max() + 1, 1)
ax.hist([courses_before_covid['Credits'], courses_during_covid['Credits'], courses_after_covid['Credits']], bins=bins, stacked=True, label=['Before COVID', 'During COVID', 'After COVID'], alpha=0.75)
ax.set_xlim(0, bins[-1])

# Customize the plot appearance
ax.set_xlabel('Course Credits', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
ax.set_title('Course Credits Distribution Before, During, and After COVID', fontsize=16)
ax.legend(title='Period', title_fontsize=14, fontsize=12)

# Remove top and right spines for a cleaner look
sns.despine()

plt.show()
```

# Another Slight Difference

It appears that there are actually more course credits taken across all course credit distributions post-COVID. During COVID we see an increase of about 12000 in the 3 credit hour category (the most common) from pre-COVID levels. It appears like there's not much information to be gathered here though, as this could be explained by a variety of factors such as increase in enrollments and changes to the curriculum.

How did the pandemic affect the GPA distribution between different subjects? Could the increase in student performance be due to different classes being taken? To determine this, we must first ask what the most common courses were taken before, during, and after COVID.

```python
subject_counts_before = courses_before_covid['Subject'].value_counts().reset_index().rename(columns={'index': 'Subject', 'Subject': 'Count'})
subject_counts_during = courses_during_covid['Subject'].value_counts().reset_index().rename(columns={'index': 'Subject', 'Subject': 'Count'})
subject_counts_after = courses_after_covid['Subject'].value_counts().reset_index().rename(columns={'index': 'Subject', 'Subject': 'Count'})

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 6), sharey=True)
sns.barplot(data=subject_counts_before.head(10), x='Subject', y='Count', ax=axes[0])
axes[0].set_title('Top 10 Subjects Before COVID')
axes[0].set_xlabel('Subject')
axes[0].set_ylabel('Course Count')

sns.barplot(data=subject_counts_during.head(10), x='Subject', y='Count', ax=axes[1])
axes[1].set_title('Top 10 Subjects During COVID')
axes[1].set_xlabel('Subject')
axes[1].set_ylabel('Course Count')

sns.barplot(data=subject_counts_after.head(10), x='Subject', y='Count', ax=axes[2])
axes[2].set_title('Top 10 Subjects After COVID')
axes[2].set_xlabel('Subject')
axes[2].set_ylabel('Course Count')

plt.tight_layout()
plt.show()
```

# Little Difference Between Subjects Taken

Between the three time periods, there is very little difference in terms of the breakdown of classes. The sole difference between during and before is that communications no longer made the top ten of class subjects taken during COVID, and was replaced by ACIS.

Something notable, however, is the decrease in total course count during and after COVID when compared to pre-COVID levels. It appears that almost all subjects had their course count halved during and post COVID. This appears to indicate that fewer courses were offered as the university consolodated the classes taught.

This brings us to our next question: on a departmental basis, what is the difference between the pre-COVID and post-COVID GPA?

```python
# Group by department and calculate mean GPA
mean_gpa_before_covid = courses_before_covid.groupby('Subject')['GPA'].mean().reset_index()
mean_gpa_during_covid = courses_during_covid.groupby('Subject')['GPA'].mean().reset_index()
mean_gpa_after_covid = courses_after_covid.groupby('Subject')['GPA'].mean().reset_index()

# Merge the results into a single dataframe
mean_gpa_all = mean_gpa_before_covid.merge(mean_gpa_during_covid, on='Subject', suffixes=('_before', '_during'))
mean_gpa_all = mean_gpa_all.merge(mean_gpa_after_covid, on='Subject')
mean_gpa_all.columns = ['Subject', 'Mean GPA Before COVID', 'Mean GPA During COVID', 'Mean GPA After COVID']

# Calculate the difference in mean GPA between before and during COVID
mean_gpa_all['GPA Difference'] = mean_gpa_all['Mean GPA During COVID'] - mean_gpa_all['Mean GPA Before COVID']

# Sort all departments by GPA Difference
all_departments_sorted = mean_gpa_all.sort_values('GPA Difference', ascending=False)

# Create a bar chart for all departments
plt.figure(figsize=(20, 10))
sns.barplot(data=all_departments_sorted, x='Subject', y='GPA Difference')
plt.title('GPA Discrepancy by Department During COVID')
plt.xlabel('Department')
plt.ylabel('GPA Difference (During COVID - Before COVID)')
plt.xticks(rotation=90)
plt.show()
```

# GPA Difference

Consistent with our previous analyses, we can see that on the whole, GPA increased during the pandemic compared to pre-pandemic. Only a few departments saw their average GPA lower.

Given that, what are the top ten departments with increases/decreases in GPA during COVID when compared to the pre-pandemic era?

```python
# Calculate the difference in mean GPA between before and during COVID
mean_gpa_all['GPA Difference'] = mean_gpa_all['Mean GPA During COVID'] - mean_gpa_all['Mean GPA Before COVID']

# Sort by GPA Difference and separate into top 10 increased and top 10 decreased
top_10_increased = mean_gpa_all.nlargest(10, 'GPA Difference')
top_10_decreased = mean_gpa_all.nsmallest(10, 'GPA Difference')

# Function to create a bar chart for the top 10 departments
def plot_top_10_departments(data, title):
    plt.figure(figsize=(12, 6))
    sns.barplot(data=data, x='Subject', y='GPA Difference')
    plt.title(title)
    plt.xlabel('Department')
    plt.ylabel('GPA Difference (During COVID - Before COVID)')
    plt.show()

# Create bar charts for the top 10 departments with increased and decreased GPA
plot_top_10_departments(top_10_increased, 'Top 10 Departments with Increased GPA During COVID')
plot_top_10_departments(top_10_decreased, 'Top 10 Departments with Decreased GPA During COVID')
```

# STEM vs. People-Centered Studies

A general trend emerges: STEM classes made up the top ten courses with GPA increases, while classes that centered around interactions among people (Hebrew, Military Science, Leadership) made up the majority of the classes that saw GPA decreases.

Now that begs the question: does this same trend continue after COVID?

```python
# Calculate the difference in mean GPA between before and after COVID
mean_gpa_all['GPA Difference After'] = mean_gpa_all['Mean GPA After COVID'] - mean_gpa_all['Mean GPA Before COVID']

# Sort all departments by GPA Difference After
all_departments_sorted_after = mean_gpa_all.sort_values('GPA Difference After', ascending=False)

# Create a bar chart for all departments
plt.figure(figsize=(20, 10))
sns.barplot(data=all_departments_sorted_after, x='Subject', y='GPA Difference After')
plt.title('GPA Discrepancy by Department After COVID')
plt.xlabel('Department')
plt.ylabel('GPA Difference (After COVID - Before COVID)')
plt.xticks(rotation=90)
plt.show()
```

There are a variety of reasons for why this may be, and with one of them surrounding the number of students at Virginia Tech during our given time periods. How did the university grow/shrink during COVID?

```python
import os

def plot_admissions_trends(csv_file, category='All'):
    # Read the CSV file
    df = pd.read_csv(os.path.join('Admissions', csv_file))
    
    # Transpose the dataframe and set column names
    df = df.T
    df.columns = df.iloc[0]
    df = df.iloc[1:]
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'Year'}, inplace=True)
    
    # Calculate average differences before and after COVID
    pre_covid = df[df['Year'] <= '2020-21']
    post_covid = df[df['Year'] >= '2021-22']
    
    pre_covid_avg = pre_covid[['Applied', 'Offered', 'Enrolled']].mean()
    post_covid_avg = post_covid[['Applied', 'Offered', 'Enrolled']].mean()
    avg_difference = post_covid_avg - pre_covid_avg

    # Calculate offered rates for pre and post COVID
    pre_covid_offered_rate = (pre_covid['Offered'].astype(float) / pre_covid['Applied'].astype(float)).mean()
    post_covid_offered_rate = (post_covid['Offered'].astype(float) / post_covid['Applied'].astype(float)).mean()


    # Plot the trends
    plt.figure(figsize=(14, 8))

    # Customize Seaborn style
    sns.set_style('darkgrid')
    sns.set_palette('tab10')

    # Line plots
    applied_plot = sns.lineplot(x='Year', y='Applied', data=df, label='Applied', marker='o', linewidth=2.5)
    offered_plot = sns.lineplot(x='Year', y='Offered', data=df, label='Offered', marker='o', linewidth=2.5)
    enrolled_plot = sns.lineplot(x='Year', y='Enrolled', data=df, label='Enrolled', marker='o', linewidth=2.5)

    # COVID vertical line (halfway between 2020-21 and 2021-22)
    covid_start_index = df[df['Year'] == '2020-21'].index[0] + 0.5
    plt.axvline(x=covid_start_index, color='red', linestyle='--', label='COVID-19 Start')
    
    # Add coordinate values
    for index, row in df.iterrows():
        plt.text(index - 0.3, row['Applied'] + 200, row['Applied'], fontsize=11, fontweight='bold')
        plt.text(index - 0.3, row['Offered'] - 600, row['Offered'], fontsize=11, fontweight='bold')
        plt.text(index - 0.3, row['Enrolled'] - 600, row['Enrolled'], fontsize=11, fontweight='bold')

    # Customize plot
    plt.xlabel('Year', fontsize=14, labelpad=15)
    plt.ylabel('Count', fontsize=14, labelpad=15)
    plt.title(f'{category} Student Admissions Trends', fontsize=18, pad=20)
    plt.xticks(df.index, df['Year'], rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    
    print(f"{category} - Pre COVID Offered Rate: {pre_covid_offered_rate:.2%}")
    print(f"{category} - Post COVID Offered Rate: {post_covid_offered_rate:.2%}")

    # Add average differences and offered rates to the legend
    legend_labels = [
        f"Applied (Î”: {avg_difference['Applied']:.2f})",
        f"Offered (Î”: {avg_difference['Offered']:.2f})",
        f"Enrolled (Î”: {avg_difference['Enrolled']:.2f})",
        'COVID-19 Start'
    ]

    handles, _ = plt.gca().get_legend_handles_labels()
    plt.legend(handles=handles, labels=legend_labels, fontsize=12)

    plt.show()
    
# Call the function with the example CSV file
plot_admissions_trends('Student_Admissions.csv', category='All')
```

# Significant Difference in Admissions

Enrollments have remained relatively static when comparing before and after COVID-19. However, total applications and admittance offerings have increased significantly, with the number of applications between the years of 2020-21 and 2021-22 increasing by roughly 11,000. Additionally, over that same course of time, the number of admittances offered increased as well, by about 3500. This informs us that COVID-19 gave Virginia Tech a significantly larger application pool to draw students from and offer admittance to.

How does this break down by student demographics? For this study, we'll be using gender, first-gen status, in-state status, international status, and race as the demographics of interest.

```python
def plot_multiple_admissions_trends(csv_files, categories, title='Student Admissions Trends'):
    plt.figure(figsize=(14, 8))
    
    # Customize Seaborn style
    sns.set_style('darkgrid')
    sns.set_palette('tab10')

    legend_labels = []
    handles = []

    for csv_file, category in zip(csv_files, categories):
        # Read the CSV file
        df = pd.read_csv(os.path.join('Admissions', csv_file))
        
        # Transpose the dataframe and set column names
        df = df.T
        df.columns = df.iloc[0]
        df = df.iloc[1:]
        df.reset_index(inplace=True)
        df.rename(columns={'index': 'Year'}, inplace=True)
        
        # Calculate average differences before and after COVID
        pre_covid = df[df['Year'] <= '2020-21']
        post_covid = df[df['Year'] >= '2021-22']
        
        pre_covid_avg = pre_covid[['Applied', 'Offered', 'Enrolled']].mean()
        post_covid_avg = post_covid[['Applied', 'Offered', 'Enrolled']].mean()
        avg_difference = post_covid_avg - pre_covid_avg
        
        # Calculate offered rate for pre and post COVID
        pre_covid_offered_rate = (pre_covid['Offered'] / pre_covid['Applied']).mean() * 100
        post_covid_offered_rate = (post_covid['Offered'] / post_covid['Applied']).mean() * 100

        # Line plots with coordinate values
        for col in ['Applied', 'Offered', 'Enrolled']:
            plot = sns.lineplot(x='Year', y=col, data=df, marker='o', linewidth=2.5)
            handles.append(plot.lines[-1])
            legend_labels.append(f"{category} - {col} (Î”: {avg_difference[col]:.2f})")
            
            for index, row in df.iterrows():
                value = row[col]
                y_offset = 200 if col == 'Applied' else -600
                plt.text(index - 0.3, value + y_offset, int(value), fontsize=11, fontweight='bold')
    
        # Add offered rate for pre and post COVID to the legend
        print(f"{category} - Pre COVID Offered Rate: {pre_covid_offered_rate:.2f}%")
        print(f"{category} - Post COVID Offered Rate: {post_covid_offered_rate:.2f}%")
    
    # COVID vertical line (halfway between 2020-21 and 2021-22)
    covid_start_index = df[df['Year'] == '2020-21'].index[0] + 0.5
    plt.axvline(x=covid_start_index, color='red', linestyle='--', label='COVID-19 Start')
    handles.append(plt.gca().lines[-1])
    legend_labels.append('COVID-19 Start')
    
    # Customize plot
    plt.xlabel('Year', fontsize=14, labelpad=15)
    plt.ylabel('Count', fontsize=14, labelpad=15)
    plt.title(title, fontsize=18, pad=20)
    plt.xticks(df.index, df['Year'], rotation=45, fontsize=12)
    plt.yticks(fontsize=12)

    # Add average differences to the legend
    plt.legend(handles=handles, labels=legend_labels, fontsize=12)

    plt.show()

csv_files = ['Male_Admissions.csv', 'Female_Admissions.csv']
categories = ['Male', 'Female']
title = 'Male vs Female Students Admissions Trends'

plot_multiple_admissions_trends(csv_files, categories, title)
```

# Gender Breakdown is Consistent with the Student Body as a Whole

Trends from before the pandemic, such as having more male applicants as well as enrollees, were maintained post-pandemic. With that said, there are some differences. Increase in female applications as a percentage of pre-pandemic levels increased far more than that of male applicants (54% vs. 45%, respectively). The gap between male and female enrollees, however, remained roughly stagnant post pandemic when compared to pre-pandemic levels.

```python
csv_files = ['First_Gen_Admissions.csv', 'Non_First_Gen_Admissions.csv']
categories = ['First_Gen', 'Non_First_Gen']
title = 'First Gen vs Non First Gen Students Admissions Trends'

plot_multiple_admissions_trends(csv_files, categories, title)
```

This hints that a significant amount of the growth in the number of enrollees at Virginia Tech came from first-gen applicants.

```python
csv_files = ['In_State_Admissions.csv', 'Out_Of_State_Admissions.csv']
categories = ['In_State', 'Out_Of_State']
title = 'In State vs Out Of State Students Admissions Trends'

plot_multiple_admissions_trends(csv_files, categories, title)
```

# Out of State Interest Far Higher Post-Pandemic

Out of state applicants, pre-pandemic, roughly equalled the number of in-state applicants. Post-COVID, this was no longer the case, as out of state saw an increase of 73% applicants when compared to pre-pandemic levels. In-state, over the same period of time, only saw an increase of 41%. This affected the overall structure of post-pandemic classes, as in-state students took up 245 fewer slots for the class of 26 compared to the class of 24, while out of state students increased by 602 students over the same time period. On the whole, Virginia Tech's student population became less Virginian after COVID-19. This is further evidenced by the difference in offer rates post-COVID: 46.8% for an in-state student, 64.6% for an out of state student.

```python
csv_files = ['International_Admissions.csv', 'Non_International_Admissions.csv']
categories = ['International', 'Non_International']
title = 'International vs Non International Students Admissions Trends'

plot_multiple_admissions_trends(csv_files, categories, title)
```

While domestic students saw an increase in applications by 71%, international interest in Virginia Tech only grew by a relatively mild 50%.
```python
plot_admissions_trends('White_Admissions.csv', category='White')
plot_admissions_trends('Black_Admissions.csv', category='Black')
plot_admissions_trends('Hispanic_Admissions.csv', category='Hispanic')
plot_admissions_trends('Asian_Admissions.csv', category='Asian')
```

# Steady Increases in Applications from All Races, But Decreases in Acceptance Rates for Some

All races (White, Black, Hispanic, and Asian) saw increases in application numbers post-COVID, but their distributions were not all the same.

### White

There was an increase of 1500 offers post-COVID to White students. This is despite White students' acceptance rate falling 14% when compared to pre-COVID rates (down from 71%). Keeping in line with the reduction in acceptance rate, however, was the enrollment: 135 fewer White students enrolled at Virginia Tech post-pandemic.


### Black

There was a significant increase in the number of Black students applying to Virginia Tech post-pandemic, with applications increasing by roughly 1900. This led to an increase in offers (1000) and enrollments (181) when compared to pre-pandemic numbers. Additionally, the overall acceptance rate for Black students increased by 2%, up from 56%.

### Hispanic

Hispanic students also saw an increase in applications, offers, and enrollments, with similar numbers to African-American students (roughly 1800 more applications, 1250 more offers, and 270 more enrollments when compared with pre-pandemic numbers). The acceptance rate of Hispanic students also increased post-COVID by 2%, up from 66%.

### Asian

Asian students had the largest increase in number of applications by minority groups, and also saw an increase in offers and enrollments (roughly 2900 more applications, 1230 more offers, and 153 more enrollments). Interestly, though, they also saw their acceptance rate decrease by 12%, down from 67%.

On the whole, it appears that COVID-19 made Virginia Tech into a much more racially and ethnically diverse university.
